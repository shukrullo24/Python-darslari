# Ro'yxatlar bilan ishlash



```
.sort() metodi
```
List ichidagi elementlarni alifbo tartibi bo'yicha tartiblaydi


````python
cars = ['bmw', 'mercedes benz', 'volvo', 'general motors', 'tesla']
print("Tartiblangamagan holatdagi mashinalar ro'yxati:", cars)
cars.sort()
print("Tartiblangan ro'yxati:", cars)
````

Ro'yxatni teskari tartiblash

````python
cars = ['bmw', 'mercedes benz', 'volvo', 'general motors', 'tesla']
cars.sort(reverse=True)
print(cars)
````

`.sort()` metodi ro'yxatni tarkibini butunlay o'zgartirib tartiblaydi.

`.sorted()` metodi esa ro'yxatni tarkibini o'zgartirmagan holatda tartiblaydi

````python
mehmonlar = ['Odil', 'Hamid', 'Temur', 'Avazbek', 'Farruh', 'Shamsiddin']

print(f"Vaqtincha tartiblangan ro'yxat: {sorted(mehmonlar)}")
print(f"Asl ro'yxat: {mehmonlar}")
````

`sorted(argument)` funksiyasi uchun ham `reverse(True)` argumentidan foydalanish mumkin

````python
mehmonlar = ['Odil', 'Hamid', 'Temur', 'Avazbek', 'Bunyodbek', 'Akbar', 'Shaxboz']
print(f"Teskari tartiblangan ro'yxat: {sorted(mehmonlar, reverse=True)}")
print(f"Bu esa asl ro'yxat: {mehmonlar}\n")
print(f"Bu tartiblangan ro'yxat: {sorted(mehmonlar)}")
print(f"Bu esa asl ro'yxat: {mehmonlar}")
````

Ro'yxatni teskari tartiblashning yana bir usuli

````python
fruits = ['pear','banana','apple','watermelon','lemon']
print("Asl ro'yxat: ", fruits)
# fruits.reverse()
# print(f"Teskari tartiblangan ro'yxat: {fruits}")

tes_royxat = reversed(fruits)
print(f"Teskari tartiblangan ro'yxat: {tes_royxat}")
````

Ro'yxatning uzunligini bilish

````python
fruits = ['pear','banana','apple','watermelon','lemon']
roy_uzunligi = len(fruits)
print("Ushbu ro'yxat", roy_uzunligi, "elementdan iborat")
````

Ro'yxatdan nusxa olish

````python
sonlar = [1, 2, 3, 4, 5] # donlar degan ro'yxat yaratamiz
sonlar2 = sonlar # sonlar2 degan ro'yxatni sonlar ga tenglaymiz
sonlar2.append(6)
sonlar2.append(7)

print("Bu sonlar ro'yxati: ", sonlar)
print("Bu sonlar2 ro'yxati: ", sonlar2)
````

Yuqoridagi usul orqali biz bitta ro'yxatni ikki o'zgaruvchiga bog'lab qo'ydik ulardan nuxsa olinmadi

````python
sonlar = [1, 2, 3, 4, 5] # sonlar ro'yxatini yaratdik
sonlar2 = sonlar[:] # [:] ro'yxatni to'liq ko'chirib/nusxalab oladi
sonlar2.append(6)
sonlar2.append(7)
print("Bu sonlar ro'yxati: ", sonlar)
print("Bu sonlar2 ro'yxati: ", sonlar2) # sonlar2 ro'yxatiga 6, 7 raqamlari qo'shildi
````

`Tuple()` o'zgarmas ro'yxatlar

````python
pi = (3.145)
print(pi) # pi o'zgarmas qiymat sifatida saqlandi

# ro'yxatga ko'proq element qo'shib ko'ramiz
toys = ['bus', 'car', 'bear', 'dino', 'snake', 'lizard']
print(toys)
print(toys[0])
print(toys[1:5])
````

````python
toys = tuple(toys)
toys[2] = 'dragon'
print(toys)
# toys[2] = 'olma'
# toys
````

Yuqoridagi natija xato qaytarishi tabiiy chunki biz ro'yxatni `tuple()` funksiyasi yordamida o'zgarmas ro'yxatga o'zgartirdik.
