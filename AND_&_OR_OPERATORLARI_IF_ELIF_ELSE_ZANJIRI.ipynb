# Bir nechta shartlarni tekshirish
### `in` operatori

in operatori yordamida biz ro'yxatning tarkibini ma'lum bir element borligini tekshirishimiz mumkin

```python
menu = ['osh','qazonkabob','shashlik','norin','somsa']
'muzqaymoq' in menu # menu da manti bormi?
```

```python
menu = ['osh','qazonkabob','shashlik','norin','somsa']
'osh' in menu # menu da osh bormi?
```

```python
menu = ['osh','qazonkabob','shashlik','norin','somsa']
taom = input("Nima ovqat buyurtma qilasiz?\n>>> ")
if taom in menu:
  print("Buyurtma qabul qilindi")
else: print(f"Afsuski bizda {taom} mavjud emas")
```

`not in` operatori

```python
menu = ['osh','qazonkabob','shashlik','norin','somsa']
taom = input("Nima ovqat buyurtma qilasiz? ")
if taom.lower() not in menu:
  print(f"{taom.title()} menuda mavjud emas")
else:
  print(f"{taom.title()} buyurtmangiz qabul qilindi")
```

`if not` operatori

```python
a = int(input("5 yoki undan katta son kiriting: "))
if not a > 5:
  print(f"{a} soni 5 dan katta emas")
else: print(f"{a} soni 5 dan katta")
```

### Ikki ro'yxatni solishtirish
Ikki ro'yxatning tarkibi quyidagicha tekshiriladi:

```python
menu = ['osh','qazonkabob','shashlik','norin','somsa']
buyurtmalar = ["osh","somsa","manti", "shashlik"]

for taom in buyurtmalar:
  if taom in menu:
    print(f"Menuda {taom} bor")
  else:
    print(f"Menuda {taom} mavjud emas")
```

Ro'yxat bo'sh emasligini tekshiramiz

```python
menu = ['osh','qazonkabob','shashlik','norin','somsa']
# buyurtmalar = ["osh","somsa","manti", "shashlik"]
buyurtmalar = []

if buyurtmalar:
  for ovqat in buyurtmalar:
    if ovqat in menu:
      print(f"Menuda {ovqat} bor")
    else: print(f"Menuda {ovqat} yo'q")
else:
  print("Savatchangiz bo'sh")
```

`mahsulotlar` degan ro'yxat yarating va kamida 10 ta turli mahsulotni kiriting. Yangi, `savat` degan bo'sh ro'yxat yarating va foydalanuvchidan savatga kamida 5 ta mahsulot kiritishni so'rang. Savatdagi elementlarni, `mahsulotlar` ro'yxati bilan solishtiring va qaysi biri ro'yxatda bo'lsa "Mahsulot do'konimizda bor" aks holda, "Mahsulot do'konimizda yo'q" degan xabarlarni chiqaring.

```python
mahsulotlar = ["pishloq", "sut", "qatiq", "suzma", "yogurt", "smetana", "muzqaymoq", "qaymoq", "sariyoq", "dolta"]
savat = []
print("Savatga kamida 5ta mahsulot kiritishingiz lozim.")
for n in range(5):
  savat.append(input(f"Marhamat {n+1}-mahsulotni kiriting: "))

print("\n--- Savatingizdagi mahsulotlarni tekshirish ---")
for mahsulot in savat:
  if mahsulot in mahsulotlar:
    print(f"{mahsulot} do'konimizda bor")
  else:
    print(f"{mahsulot} do'konimizda mavjud emas")
```
