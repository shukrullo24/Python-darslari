Parameters in function

```python
# Greet employees by name
def greet_employee(name):
  print("Salom, xush kelibsiz", name.title())
greet_employee("shukrullo") # Shukrullo is agrument sent to the parameter
```

```python
# Greet employees by name
def greet_employee(name, surname):
  print("Salom, xush kelibsiz", name.title(), surname.title())
greet_employee("shukrullo", "soginboev")
```

```python
# Return information from a function
def calculate_fails(total_attempts, failed_attempts):
  fail_percentage = failed_attempts / total_attempts * 100
  return fail_percentage

percentage = calculate_fails(4, 2)
if percentage >= 50:
  print("Account locked for a day")
```
