# Sonlar

> Pythonda sonlar bilna ishlash bo'yicha amaliy mashqlar


Quyidagi dasturlarning har birini alohida fayl ko'rinishida yozing va bajaring:

Foydalanuvchi kiritgan sonning kvadrati va kubini konsolga chiqaruvchi dastur yozing

```python
son_1 = int(input("Istalgan son kiriting, uning kvadrati va kubini hisoblab beraman: "))
kv = son_1**2
kub = son_1**3

natija = "Kiritilgan sonning kvadrati ", kv, "ga teng, kubi esa ", kub, "ga teng"
print(natija)
```

Foydalanvuchining yoshini so'rab uning tug'ilgan yilini konsolga chiqaruvchi dastur

```python
yosh = int(input("Yoshingizni kiriting: "))
yil = 2026
natija = yil - yosh
print(f"Siz {natija}-yilda tug'ilgnasiz")
```

Foydalanuvchidan ikki son kiritishini so'rab, sonlarning yig'indisi, ayirmasi, ko'paytmasi va bo'linmasini chiqaruvchi dastur

```python
print("Kiritilgan sonlarning yig'indisi, ayirmasi, ko'paytmasi va bo'linmasini hisoblaymiz")
num_1 = int(input("Birinchi sonni kiriting: "))
num_2 = int(input("Ikkinchi sonni kiriting: "))

yigindi = num_1 + num_2
ayirma = num_1 - num_2
kopaytma = num_1 * num_2
bolinma = num_1 / num_2

natija_1 = [yigindi, ayirma, kopaytma, bolinma]
natija_2 = f"Kiritilgan sonlarning yig'indisi {yigindi}, ayirmasi {ayirma}, ko'paytmasi {kopaytma}, bo'linmasi {bolinma} ga teng."
print(natija_1)
print(natija_2)
```

# List (Ro'yxat)

`ismlar` degna ro'yxat yarating va kamida 3 ta yaqin do'stingizning ismini kiriting.
Ro'yxatdagi har bir do'stingizga qisqa xabar yozib konsolga chiqaring:
"Salom Abror, bugun choyxona bormi?
Mahmud, choyxonaga boramizmi?"



```python
ismlar = ['oybek', 'maqsud', 'islom']
print(f"Salom {ismlar[0].title()}, bugun choyxona bormi?")
print(f"Salom {ismlar[1].title()}, bugun choyxona bormi?")
print(f"Salom {ismlar[2].title()}, bugun choyxona bormi?")
```

`sonlar` deb nomlangan ro'yxat yarating va ichiga turli sonlarni yuklang (musbat, manfiy, butun, o'nlik)

```python
sonlar = [10, -20, 35, 12.5]
```

Yuqoridagi sonlar ustida turli arifmetik amallar bajarib ko'ring.
Ro'yxatdagi ba'zi sonlarning qiymatini o'zgartiring, ba'zilarini esa almshtiring

```python
sonlar = [10, -20, 35, 12.5]
sonlar.remove(10)
print(sonlar)
```

```python
sonlar = [10, -20, 35, 12.5]
sonlar.remove(10)
print(sonlar)
sonlar.insert(0, 25)
print(sonlar)
```
