# Ro'yxatlar bilan ishlash bo'yicha amalyot

O'zingizga ma'lum davlatlarning ro'yxatini tuzing va ro'yxatni konsolga chiqaring

```python
davlatlar = ['O\'zbekiston', 'Janubiy Korea', 'Turkiya', 'Buyuk Britaniya', 'Canada']
davlatlar
```

Ro'yxatning uzunligini konsolga chiqaring

```python
dav_soni = len(davlatlar)
print("Ro'yxatimizdagi davlar soni ", dav_soni, "ta")
```

`sorted()` funktsiyasi yordamida ro'yxatni tartiblangan holda konsolga chiqaring

```python
print("Tartiblangan ro'yxat: ", sorted(davlatlar))
print("Asl ro'yxat: ", davlatlar)
```

`sorted()` yordamida ro'yxatni teskari tartibda konsolga chiqaring

```python
print("Teskari tartiblangan ro'yxat: ", sorted(davlatlar, reverse=True))
print("Asl ro'yxat: ", davlatlar)
```

Asl ro'yxatni qaytadan konsolga chiqaring

```python
print("Asl ro'yxat: ", davlatlar)
```

`reverse()` metodi yordamida ro'yxatni ortidan boshlab chiqaring

```python
davlatlar.reverse()
davlatlar
```

`sort()` metodi yordamida ro'yxatni avval alifbo bo'yicha, keyin esa alifboga teskari tartibda konsolga chiqaring.

```python
davlatlar.sort()
print(davlatlar)
sorted(davlatlar, reverse=True)
davlatlar
```

120 dan 1200 gacha bo'lgan juft sonlar ro'yxatini tuzing

```python
juft_sonlar = list(range(120, 1201, 2))
juft_sonlar
```

Ro'yxatdagi sonlar yig'indisini hisoblang va konsolga chiqaring

```python
print(sum(juft_sonlar))
```

Ro'yxatdagi eng katta va eng kichik son o'rtasidagi ayirmani hisoblang va natijani konsolga chiqaring

```python
ayirma = max(juft_sonlar) - min(juft_sonlar)
print("Ro'yxatdagi eng katta sondan eng kichik sonning ayirmasi", ayirma, "ga teng")
```

Ro'yxatdagi elementlar sonini sonini hisoblang

```python
len(juft_sonlar)
```

`taomlar` degan ro'yxat yarating hamda ichiga istalgan 5ta taomni kiriting

`nonushta` degan yangi ro'yxatga `taomlar`dan nusxa oling

```python
taomlar = ['osh', 'manti', 'somsa', 'chuchvara', 'xonim']
nonushta = taomlar[:]
print(f"Taomlardagi ro'yxat: {taomlar}")
print(f"Nonushtadagi toamlar ro'yxati: {nonushta}")
```

Yangi ro'yxatda faqat nonushtaga yeyiladigan taomlarni qoldiring, va qo'shimcha 2 ta taom qo'shing

```python
taomlar = ['osh', 'manti', 'somsa', 'chuchvara', 'xonim']
nonushta.remove("xonim")
nonushta.remove("manti")
nonushta.remove("chuchvara")
nonushta.append("mastava")
nonushta.append("sho'rva")
nonushta
```

Ikkala ro'yxatni ham (taomlar va nonushta) konsolga chiqaring

```python
print("Taomlar ro'yxati:", taomlar)
print("Nonushta ro'yxati:" ,nonushta)
```
