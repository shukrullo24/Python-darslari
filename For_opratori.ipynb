`for` operatori bilan tanishamiz

```python
mehmonlar = ['ali', 'vali', 'hasan', 'husan', 'olim']
for mehmon in mehmonlar:
  print(mehmon)
```

```python
mehmonlar = ['ali', 'vali', 'hasan', 'husan', 'olim']
for mehmon in mehmonlar:
  print(f"Hurmatli {mehmon}, sizni 20-dekabr kungi oshga taklif etamiz \nHurmat bilan, Palonchiyevlar oilasi")
```

Sonli ro'yxatlar bilan ishlash

```python
sonlar = list(range(1, 21, 2))
for son in sonlar:
  print(f"{son} ning kvadrati {son**2}, kubi esa {son**3} ga teng")
```

for yordamida ro'yxat shakllantirish

```python
sonlar = list(range(11))
sonlar_kvadrati = []
for son in sonlar:
  sonlar_kvadrati.append(son**2)
print(sonlar)
print(sonlar_kvadrati)
```

for operati va input()

```python
dostlar = []
print("5 ta eng yaqin do'stingiz kim?")
for n in range(5):
  dostlar.append(input(f"{n+1}-do'stingiz ismini kiriting: "))
dostlar = str(dostlar)
print(dostlar.title())
```
