`if` operatori

```python
avtolar = ['audi', 'bmw', 'volvo', 'kia', 'hyundai']
avtolar
```

```python
for avto in avtolar:
  if avto == "bmw":
    print(avto.upper())
  else:
    print(avto.title())
```
